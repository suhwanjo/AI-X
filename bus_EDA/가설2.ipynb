# 연령별생활인구 데이터 정리

```python
a=pd.read_csv('연령별인구.csv',sep=',',encoding='cp949')
```

```python
a.head()
```

```python
a['기준일ID'] = pd.to_datetime(a['기준일ID'], format='%Y%m%d')
a.set_index('기준일ID',inplace=True)
a.head()
```

```python
# 1.2에서 22년 4월 데이터를 사용했기 때문에 4월 데이터만 추출
b=a['2022-04-01':'2022-04-30']

b['자치구코드'].unique()
# 자치구코드가 과제와 다르게 표기된 것으로 보임
```

```python
b=b.reset_index()
b['자치구코드']=b['자치구코드'].map({
    11110:'종로구',
    11140:'중구',
    11170:'용산구',
    11200:'성동구',
    11215:'광진구', 
    11230:'동대문구',
    11260:'중랑구',
    11290:'성북구',
    11305:'강북구',
    11320:'도봉구',
    11350:'노원구',
    11380:'은평구',
    11410:'서대문구',
    11440:'마포구',
    11470:'양천구',
    11500:'강서구',
    11530:'구로구',
    11545:'금천구',
    11560:'영등포구',
    11590:'동작구',
    11620:'관악구',
    11650:'서초구',
    11680:'강남구',
    11710:'송파구',
    11740:'강동구'})
b.rename(columns={'자치구코드':'자치구'},inplace=True)
b.head()
```

```python
# 1.2에서 낮 시간 25~30세의 유동인구가 많은 것을 확인
c=b[['시간대구분','자치구','총생활인구수','남자25세부터29세생활인구수','여자25세부터29세생활인구수']]
```

```python
c.head()
```

```python
c.info()
```

```python
# 1. 25~30세 생활인구 비율이 높은 구의 정류장,노선 수가 많을 것이다.
d=c.query('시간대구분 !=[0,1,2,3,4]')
# 2. 출퇴근 시간의 평균생활인구가 많은 구의 정류장,노선 수가 많을 것이다.
h=c.query('시간대구분==[8,18]')
```

```python
e=d.groupby('자치구')[['총생활인구수','남자25세부터29세생활인구수','여자25세부터29세생활인구수']].sum().reset_index()

i=h.groupby('자치구')['총생활인구수'].mean().reset_index()
```

# 정류장, 노선 수 데이터 정리

```python
f=pd.read_csv('df_seoul_bus_station.csv')
```

```python
f.head()
```

```python
f['승하차평균승객수']=f['승차평균승객수']+f['하차평균승객수']
```

```python
g=pd.merge(e,f,how='inner',on='자치구')

j=pd.merge(i,f,how='inner',on='자치구')
```

```python
g['25세부터29세생활인구비율']=(g['남자25세부터29세생활인구수']+g['여자25세부터29세생활인구수'])/g['총생활인구수']
```

```python
g=g[['자치구','25세부터29세생활인구비율','정류장수','노선수']].set_index('자치구')

j=j[['자치구','총생활인구수','승하차평균승객수','정류장수','노선수']].set_index('자치구')
```

```python
g.head()
```

```python
j.head()
```

```python
j=j.reset_index()
```

```python
j.to_csv("commute.csv", index = False)
```

```python
# 시각화, 한글폰트 설정
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid') 
```

```python
# 1. 25~30세 생활인구 비율이 높은 구의 정류장,노선 수가 많을 것이다.
print(stats.pearsonr(g['25세부터29세생활인구비율'],g['정류장수']))
print(stats.pearsonr(g['25세부터29세생활인구비율'],g['노선수']))
# 정류장 가설 P > 0.05로 무의미
# 노선 가설 P < 0.05로 유의미

# 정류장 가설 기각
```

```python
plt.scatter('25세부터29세생활인구비율','노선수',data=g)
for name,x,y in zip(g.index,g['25세부터29세생활인구비율'],g['노선수']): 
    plt.text(x, y, name,fontsize=8) 
plt.show()
```

```python
(g['노선수']/g['25세부터29세생활인구비율']).sort_values(ascending=True)
```

```python
# 2. 출퇴근 시간의 평균생활인구가 많은 구의 정류장,노선 수가 많을 것이다.
print(stats.pearsonr(j['총생활인구수'],j['정류장수']))
print(stats.pearsonr(j['총생활인구수'],j['노선수']))
# 정류장 가설 P < 0.05로 유의미
# 노선 가설 P > 0.05로 무의미

# 노선 가설 기각
```

```python
plt.scatter('총생활인구수','정류장수',data=j)
for name,x,y in zip(j.index,j['총생활인구수'],j['정류장수']): 
    plt.text(x, y, name,fontsize=8) 
plt.show()
```

```python
(j['정류장수']/j['총생활인구수']).sort_values(ascending=True)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
X=j.총생활인구수.values
Y=j.정류장수.values

X=X.reshape(-1,1)
Y=Y.reshape(-1,1)

lr.fit(X, Y)
```

```python
import matplotlib.pyplot as plt
y2 = lr.coef_[0]*X + lr.intercept_
plt.scatter(X, Y)
for i in range(len(j)): # 행 개수만큼 순회
     row = j.iloc[i] # 한 행씩 꺼내기
     name = row['자치구'] # 이름이 저장된 열
     x = row['총생활인구수'] # x좌표가 저장된 열
     y = row['정류장수'] # y좌표가 저장된 열

     plt.text(x, y, name)
plt.plot(X, y2, color='red')
plt.title('y = {}*x + {}'.format(lr.coef_[0], lr.intercept_))

plt.show()
```

```python
j["출퇴근생활인구예측값오차"] = j['정류장수']-( lr.coef_[0]* j['총생활인구수'] + lr.intercept_)

new_df=j.sort_values("출퇴근생활인구예측값오차").reset_index(drop=True)

plt.figure(figsize=(20,10))
plt.title('정류장수 예측값과 오차')
plt.bar(new_df['자치구'], new_df["출퇴근생활인구예측값오차"])



plt.show()
```
